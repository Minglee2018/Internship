```python
# thay thế nhưng từ tiếng anh bằng hàm clear_row_EN
```

```python

def clear_row_EN(s):
    a =''
    for index,items in enumerate(s):
        if items.find('District') != -1:
            items = items.replace('District','')
            items = 'Quận'+items
        if items.find('Street') != -1:
            items = items.replace('Street','')
            items = 'Đường'+items
        if items.find('City') != -1:
            items = items.replace('City','')
            items = 'Thành Phố'+items
        if items.find('Road') != -1:
            items = items.replace('Road','')
            items = 'Đường'+items
        if items.find('Province') != -1:
            items = items.replace('Province','')
            items = 'Thành phố'+items
        if items.find('Building') != -1:
            items = items.replace('Building','')
            items = 'Tòa Nhà'+items
        items= items.strip()
        if index <len(s)-1:
            a+=items+','
        else:
            a+=items
    return a
```

```python
# vì file lớn e ko load được trên file notebook nên e chỉ sử lý và in ra 100 dòng đầu tiên của file labels.json
```

```python
for i in range(100):
    s = labels.location[i].split(',')
    s =  clear_row_EN(s)
    s = s.replace(',,',',').replace(',,,',',')
    labels.location[i] = s
```

```python
def duplicate_row(s):
    r = 0; a = s.split(',');st=''
    for item in a:
        if item == 'Thành Phố Hà Nội':
            r+=1; 
    for i,j in enumerate(a):
        if a[i]=='Thành Phố Hà Nội' and r>1:
            del a[int(i)]; r-=1
    a.pop()
    for item in a:
        st+=item+','
    return st
```

```python
def clear_row(p):
    a =''
    s = p.split(',') 
    for index,items in enumerate(s):
        if items.find('hoang mai')!=-1:
            items = items.replace('hoang mai','Quận Hoàng Mai')
        if items.find('Hoàng Mai')!=-1:
            items = items.replace('Hoàng Mai','Quận Hoàng Mai')
        if items.find('hoàng mai')!=-1:
            items = items.replace('hoàng mai','Quận Hoàng Mai')
        if items.find('Hoang Mai')!=-1:
            items = items.replace('Hoang Mai','Quận Hoàng Mai')
        items=items.strip()
        if index <len(s)-1:
            a+=items+','
        else:
            a+=items
    return a
```

```python
for i in range(100):
    labels.location[i] = clear_row(labels.location[i])
```
